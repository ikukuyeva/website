Purpose: Create Venn diagram for website
- Image based on example here: https://towardsdatascience.com/how-to-create-and-beautify-venn-diagrams-in-python-331129bd4ed3 and https://towardsdatascience.com/how-to-create-and-customize-venn-diagrams-in-python-263555527305

```python
from matplotlib_venn import venn2, venn2_circles, venn2_unweighted
from matplotlib_venn import venn3, venn3_circles
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
Product=[1, 2, 3]
Market=[3, 4, 5]
Fit=[3, 6, 7]
```

```python
font_size = 26
vd3=venn3(
    [set(Product),set(Market),set(Fit)],
    set_labels=('Products/Services', 'Customers Want', 'Willing to Pay For'),
    set_colors=('#e5e5e5', '#e5e5e5','#e5e5e5'), 
    alpha = 0.8)
c=venn3_circles([set(Product), set(Market),set(Fit)], 
                linestyle='-.', 
                linewidth=2, 
                color='grey')
for text in vd3.set_labels:
    text.set_fontsize(font_size);
for text in vd3.subset_labels:
    text.set_fontsize(font_size)
vd3.get_patch_by_id('111').set_color('#009999')
# Remove counts from diagram, per 
# https://stackoverflow.com/questions/58010265/how-do-you-remove-data-labels-in-matplotlib-venn
for idx, subset in enumerate(vd3.subset_labels):
    vd3.subset_labels[idx].set_visible(False)
# Annotate product market fit:
plt.annotate('Product market fit', 
             xy=vd3.get_label_by_id('111').get_position(), 
             xytext=(160,-70),
             ha='center', 
             textcoords='offset points', 
             fontsize=font_size,
             bbox=dict(boxstyle='round,pad=0.5', fc='#009999', alpha=0.8),
arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0.5',color='black', linewidth=2))
# plt.show()

# Save figure in bigger size, per:
# https://www.kite.com/python/answers/how-to-save-a-matplotlib-figure-as-a-certain-size-in-python
figure = plt.gcf()
plt.gca().set_axis_off()
plt.subplots_adjust(top = 1, bottom = 0, right = 1, left = 0, 
            hspace = 0, wspace = 0)
plt.margins(0,0)
plt.gca().xaxis.set_major_locator(plt.NullLocator())
plt.gca().yaxis.set_major_locator(plt.NullLocator())
figure.set_size_inches(6, 6)
# Remove whitespace around figure, per: 
# https://stackoverflow.com/questions/11837979/removing-white-space-around-a-saved-image-in-matplotlib
plt.savefig('../Venn-diagram.png', pad_inches=0, bbox_inches='tight')
```
